## **Análise de Sentimentos**
---



Análise de sentimentos, também conhecida como mineração de opiniões, é uma técnica de PLN que consiste em extrair informações subjetivas de textos, como opiniões, sentimentos e emoções.

O objetivo principal da análise de sentimentos é identificar a atitude expressa em um texto, que pode ser positiva, negativa ou neutra. Isso é feito por meio de algoritmos de aprendizado de máquina que examinam o texto em busca de palavras e frases que indiquem uma emoção ou uma opinião.

Essa técnica tem sido amplamente utilizada em diversas áreas, como marketing, mídias sociais, serviços ao cliente, políticas públicas, entre outras, para avaliar a percepção do público em relação a produtos, serviços, marcas, eventos, políticas etc. A análise de sentimentos pode ajudar a identificar tendências, pontos fortes e fracos, bem como direcionar a tomada de decisão.

A análise de sentimentos oferece diversas vantagens, tais como:

   * Economia de tempo e recursos: a análise de sentimentos automatizada permite processar grandes quantidades de dados em tempo real, o que economiza tempo e recursos em comparação com as análises manuais.

   * Melhor compreensão do público: a análise de sentimentos pode ajudar as empresas a entender melhor as opiniões e sentimentos do público em relação aos seus produtos, serviços, marcas ou campanhas de marketing.

   * Identificação de problemas: a análise de sentimentos pode ajudar a identificar rapidamente problemas e pontos fracos em produtos, serviços ou marcas, permitindo que as empresas tomem medidas corretivas de forma rápida.

   * Melhor engajamento com o público: a análise de sentimentos pode ajudar as empresas a melhorar seu engajamento com o público, respondendo a feedbacks e reclamações de forma rápida e eficiente.

   * Melhor tomada de decisão: a análise de sentimentos pode ajudar na tomada de decisão, fornecendo insights importantes sobre a percepção do público em relação a produtos, serviços ou políticas públicas.

   * Maior eficiência na segmentação do público: a análise de sentimentos pode ajudar a segmentar o público de forma mais eficiente, permitindo que as empresas criem campanhas de marketing mais direcionadas e relevantes para grupos específicos de consumidores.

Embora a análise de sentimentos ofereça muitas vantagens, há também alguns desafios e problemas que precisam ser considerados, tais como:

   * Ambiguidade: a linguagem natural pode ser ambígua e variar de acordo com o contexto, o que pode dificultar a identificação precisa do sentimento expresso em um texto.

   * Ironia e sarcasmo: a identificação de ironia e sarcasmo em um texto é um desafio para os algoritmos de análise de sentimentos, já que essas expressões podem indicar um sentimento oposto ao que as palavras literais sugerem.

   * Vocabulário limitado: os algoritmos de análise de sentimentos são baseados em dicionários de palavras e frases, o que significa que o vocabulário utilizado na análise pode ser limitado e não cobrir todas as nuances da linguagem natural.

   * Variação regional: as variações regionais na linguagem natural podem afetar a análise de sentimentos, já que diferentes regiões podem ter diferentes formas de expressar sentimentos ou opiniões sobre algo.

   * Dados desestruturados: a análise de sentimentos pode ser desafiadora quando os dados são desestruturados e apresentados de forma não padronizada, como em comentários em mídias sociais ou em conversas informais.

   * Viés de treinamento: os algoritmos de análise de sentimentos são treinados com dados existentes e podem refletir os preconceitos e limitações desses dados, o que pode afetar a precisão e a qualidade da análise.

   * Privacidade e ética: a análise de sentimentos pode levantar questões de privacidade e ética, já que envolve a coleta e análise de dados pessoais dos usuários.

**TextBlob**

Aqui está um exemplo simples de análise de sentimentos em inglês usando o pacote `textblob` em Python:

```
from textblob import TextBlob

texto = "I love this movie, it's amazing!"

blob = TextBlob(texto)

sentimento = blob.sentiment.polarity

if sentimento > 0:
    print("O sentimento é positivo")
elif sentimento == 0:
    print("O sentimento é neutro")
else:
    print("O sentimento é negativo")
```

Suponha que temos uma lista de avaliações em inglês sobre um produto e queremos calcular a polaridade do sentimento em cada avaliação. Podemos usar o seguinte código:

```
from textblob import TextBlob

# lista de avaliações em inglês
reviews = ['This product is amazing!',
           'I was really disappointed with this product.', 
           'The product is average, nothing special.']

# função para realizar a análise de sentimento usando TextBlob
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment_polarity = blob.sentiment.polarity
    return sentiment_polarity

# calcular a polaridade do sentimento para cada avaliação
sentiments = []
for review in reviews:
    sentiment = analyze_sentiment(review)
    sentiments.append(sentiment)

# imprimir a polaridade do sentimento de cada avaliação
for i in range(len(reviews)):
    print("Avaliação: ", reviews[i])
    print("Polaridade do Sentimento: ", sentiments[i])
```

Aqui está um exemplo de análise de sentimentos de um conjunto de avaliações de restaurantes:

```
from textblob import TextBlob

# avaliações de restaurantes
avaliacoes = [
    "The food was amazing! Best restaurant experience ever.",
    "I was disappointed with my meal. The flavors didn't come together well.",
    "I would not recommend this restaurant. Terrible service and the food was cold.",
    "The restaurant was okay. Not the best, but not the worst either.",
    "I absolutely loved the food! Would definitely come back again."
]

# calcular a polaridade média do sentimento das avaliações
polaridades = []
for avaliacao in avaliacoes:
    blob = TextBlob(avaliacao)
    polaridades.append(blob.sentiment.polarity)

polaridade_media = sum(polaridades) / len(polaridades)

# imprimir resultado da análise de sentimentos
if polaridade_media > 0:
    print("As avaliações têm um sentimento positivo.")
elif polaridade_media < 0:
    print("As avaliações têm um sentimento negativo.")
else:
    print("As avaliações têm um sentimento neutro.")
```

```
from textblob import TextBlob

# lista de avaliações em inglês
avaliacoes = ["This movie was amazing!",
              "The acting was terrible and the plot was boring.",
              "I loved the music and the cinematography was stunning.",
              "I couldn't stand the main actor and the special effects were terrible.",
              "This is the best book I've ever read.",
              "I found the writing to be dull and the characters uninteresting.",
              "The story was captivating and I couldn't put the book down.",
              "The author's writing style was tedious and the pacing was slow.",
              "This restaurant has the best food I've ever tasted.",
              "The service was terrible and the atmosphere was uninviting.",
              "The ambiance was perfect and the staff was very attentive.",
              "I would never recommend this place to anyone.",
              "The customer service was excellent and the prices were reasonable.",
              "The product is exactly what I was looking for.",
              "I was disappointed with the quality and it didn't meet my expectations."]

# contador de avaliações positivas e negativas
positivas = 0
negativas = 0

# análise de sentimentos
for avaliacao in avaliacoes:
    blob = TextBlob(avaliacao)
    polaridade = blob.sentiment.polarity
    
    if polaridade > 0:
        positivas += 1
    elif polaridade < 0:
        negativas += 1

# mostrar resultados
print(f"Total de avaliações positivas: {positivas}")
print(f"Total de avaliações negativas: {negativas}")

if positivas > negativas:
    print("A maioria das avaliações é positiva.")
elif positivas < negativas:
    print("A maioria das avaliações é negativa.")
else:
    print("Não há maioria clara de avaliações positivas ou negativas.")
```

A propriedade `subjectivity` em `TextBlob` representa o grau de objetividade ou subjetividade de um texto, sendo um valor entre 0 (totalmente objetivo) e 1 (totalmente subjetivo). Segue abaixo um exemplo de código que utiliza essa propriedade:

Suponha que você tenha uma grande quantidade de comentários em inglês sobre um produto e deseja identificar as palavras mais comuns e frequentes usadas nos comentários positivos e negativos.

Primeiro, você pode coletar os comentários e classificá-los em positivos ou negativos usando uma biblioteca de análise de sentimentos como `TextBlob`. Em seguida, pode processar os comentários para extrair as palavras mais frequentes em cada categoria.

Para realizar esse processo, você pode seguir os seguintes passos:

   * Coletar e classificar os comentários em positivos e negativos usando uma biblioteca de análise de sentimentos como `TextBlob`.
   * Pré-processar os comentários para remover pontuações, transformar todas as palavras em minúsculas e remover as chamadas "stop words", que são palavras comuns em um idioma que geralmente não contribuem para o significado do texto, como "a", "um", "e", etc.
   * Usar a biblioteca `NLTK` para tokenizar os comentários, ou seja, dividir o texto em palavras individuais.
   * Contar as palavras mais comuns em cada categoria usando a biblioteca `Counter` do Python.

Aqui está um exemplo de código que realiza esse processo:

```
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```
from textblob import TextBlob
import nltk
from nltk.corpus import stopwords
from collections import Counter
import string

# Coletar comentários e classificá-los em positivos e negativos usando TextBlob
comments = [
    "I absolutely love this product! It works great and has made my life so much easier.",
    "This product is terrible. It doesn't work at all and is a complete waste of money.",
    "I'm not sure how I feel about this product. It seems to work okay, but there are some things I don't like about it.",
    "This is the best product ever! I don't know how I ever lived without it.",
    "I had high hopes for this product, but it was a huge disappointment. It didn't work as advertised and I ended up returning it.",
    "This product is amazing! I love it so much.",
    "I'm very satisfied with this purchase. It works perfectly.",
    "Not good at all. The quality is poor and it broke after one use.",
    "Do not waste your money on this. Terrible product.",
    "I was disappointed with this product. It did not meet my expectations.",
    "This is the worst product I have ever used. It's a complete waste of money."
]

positive_comments = []
negative_comments = []

for comment in comments:
    blob = TextBlob(comment)
    if blob.sentiment.polarity > 0:
        positive_comments.append(comment)
    else:
        negative_comments.append(comment)

# Pré-processar os comentários para remover pontuações, transformar todas as palavras em minúsculas e remover stop words
def preprocess(comment):
    comment = comment.lower()
    comment = comment.translate(str.maketrans('', '', string.punctuation))
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(comment)
    words = [word for word in words if not word in stop_words]
    return words

positive_words = []
negative_words = []

# Processar os comentários para extrair as palavras mais frequentes em cada categoria
for comment in positive_comments:
    words = preprocess(comment)
    positive_words += words

for comment in negative_comments:
    words = preprocess(comment)
    negative_words += words

# Contar as palavras mais comuns em cada categoria
positive_counts = Counter(positive_words)
negative_counts = Counter(negative_words)

print("Palavras mais frequentes em comentários positivos:")
print(positive_counts.most_common(5))

print("\nPalavras mais frequentes em comentários negativos:")
print(negative_counts.most_common(5))
```

**Transformers**

Novos modelos de aprendizado profundo (*deep learning*) são introduzidos em uma taxa crescente e às vezes é difícil acompanhar todas as novidades. Dito isso, um modelo de rede neural específico provou ser especialmente eficaz para tarefas comuns de PLN. O modelo é chamado de **Transformer**.

Um **Transformer** é um modelo de aprendizado profundo que adota o mecanismo de autoatenção, ponderando diferencialmente o significado de cada parte dos dados de entrada.

A autoatenção é um mecanismo de atenção que relaciona diferentes posições de uma única sequência para computar uma representação da mesma sequência. O mecanismo de atenção fornece contexto para qualquer posição na sequência de entrada.

A motivação é que a rede deve dedicar mais foco às partes pequenas, mas importantes, dos dados. Aprender qual parte dos dados é mais importante que outra depende do contexto. O efeito melhora algumas partes dos dados de entrada enquanto diminui outras partes.

A biblioteca `transformers` fornece a funcionalidade para criar e usar modelos compartilhados. A plataforma **Hugging Face Hub** contém milhares de modelos pré-treinados que qualquer pessoa pode baixar e usar. Você também pode enviar seus próprios modelos para o Hub!

> https://huggingface.co/docs/hub/index

O objeto mais básico da biblioteca `transformers` é a função `pipeline()`. Ele conecta um modelo com suas etapas necessárias de pré-processamento e pós-processamento, permitindo inserir diretamente qualquer texto e obter uma resposta:

Aqui está um exemplo de como usar a biblioteca `transformers` da Hugging Face para realizar análise de sentimentos em uma frase em inglês:

```
!pip install transformers
```

```
from transformers import pipeline

# Carregando o modelo de análise de sentimentos
classifier = pipeline('sentiment-analysis')

# Frase de exemplo
text = "I really love this product, it's amazing!"

# Realizando a análise de sentimentos na frase
result = classifier(text)

# Extraindo a polaridade do sentimento
sentiment = result[0]['label']

# Imprimindo o resultado
print(f"A frase '{text}' tem sentimento {sentiment}.")
```

```
from transformers import pipeline

# Carregar o modelo de análise de sentimento pré-treinado
classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')

# Exemplo de texto a ser classificado
text = "I really love this product, it's amazing!"

# Executar a classificação de sentimento no texto de exemplo
result = classifier(text)[0]

# Imprimir o resultado
print(f"Sentimento: {result['label']} (Score: {result['score']:.2f})")
```
