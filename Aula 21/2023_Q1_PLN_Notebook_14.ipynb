A **Sumarização de Textos** é o processo de resumir um texto longo ou complexo em um texto mais curto, preservando as informações mais importantes e relevantes. A técnica é utilizada para facilitar a leitura e compreensão de um texto, economizando tempo e esforço.

A sumarização de texto é um processo que visa extrair ou gerar automaticamente um resumo conciso e informativo de um texto longo, como um artigo, um relatório ou uma notícia. 

Existem dois tipos principais de sumarização de textos: **sumarização extrativa** e **sumarização abstrativa**.

image.png

A **sumarização extrativa** é a técnica mais comum e mais fácil de implementar. Ela consiste em identificar as frases ou trechos mais importantes do texto original e usá-los para criar um resumo conciso. Essa técnica é chamada de `"extrativa"` porque as frases selecionadas são extraídas diretamente do texto original. As frases selecionadas geralmente são aquelas que contêm informações relevantes e representam as principais ideias do texto.

Em outras palavras, a sumarização extrativa envolve a extração direta de informações relevantes do texto original e sua apresentação em um formato mais conciso. É uma abordagem muito utilizada em aplicações de PLN, como resumos automáticos de notícias, artigos ou documentos técnicos.

Uma das vantagens da sumarização extrativa é que ela mantém a fidelidade do conteúdo original, o que significa que o resumo produzido é uma representação fiel do texto original. No entanto, a desvantagem é que, muitas vezes, a sumarização extrativa pode produzir resumos que são menos fluidos ou coesos do que os resumos produzidos por outras técnicas, como a sumarização abstrativa.


A **sumarização abstrativa** é uma técnica mais avançada e desafiadora, que visa gerar um resumo novo, com base nas informações contidas no texto original, mas expressas de forma diferente. Essa técnica é chamada de `"abstrativa"` porque requer a compreensão semântica do texto para identificar as ideias principais e expressá-las de forma mais concisa. A sumarização abstrativa é mais sofisticada do que a sumarização extrativa e é mais difícil de ser realizada com precisão, mas pode gerar resumos mais naturais e informativos.

A sumarização abstrativa envolve a compreensão do significado do texto original e a criação de um resumo em suas próprias palavras, mantendo as informações mais importantes e relevantes. Isso pode incluir a reorganização de frases, a fusão de informações de várias partes do texto e a geração de novas frases para capturar o significado geral do texto original.

Uma das principais vantagens da sumarização abstrativa é que ela pode produzir resumos mais coesos e naturais do que a sumarização extrativa, já que ela é capaz de gerar frases e parágrafos que não estão presentes no texto original, mas que ainda capturam o sentido geral do conteúdo original. No entanto, a desvantagem é que a qualidade da sumarização abstrativa depende da qualidade dos modelos de linguagem e da capacidade do algoritmo em compreender o conteúdo do texto original e gerar um resumo preciso e coeso.

Ambas as técnicas de sumarização são importantes e úteis em diferentes contextos, dependendo das necessidades do usuário e do objetivo da aplicação.

**`Frequência das palavras`**

Aqui está um exemplo simples de como realizar a **sumarização extrativa** considerando a frequência das palavras no texto e usando a biblioteca `NLTK`:

```
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# Definir o texto original a ser resumido
texto_original = "O futebol é um esporte de equipe jogado com uma bola redonda. \
É jogado por dois times de onze jogadores cada um, e o objetivo do jogo é marcar \
mais gols que o adversário, colocando a bola na meta do time adversário. \
O futebol é o esporte mais popular do mundo, com milhões de pessoas assistindo \
e jogando regularmente."

# Tokenizar o texto em frases e palavras
frases = sent_tokenize(texto_original)
palavras = word_tokenize(texto_original)

# Remover as stopwords (palavras sem significado importante para a análise)
stopwords = set(stopwords.words('portuguese'))
palavras_sem_stopwords = [palavra for palavra in palavras if not palavra.lower() in stopwords]

# Calcular a frequência de cada palavra no texto
frequencia_palavras = nltk.FreqDist(palavras_sem_stopwords)

# Identificar as frases mais importantes
frases_importantes = []
for frase in frases:
  pontuacao_total = 0
  for palavra in word_tokenize(frase.lower()):
    if palavra in frequencia_palavras:
      pontuacao_total += frequencia_palavras[palavra]
  frases_importantes.append((frase, pontuacao_total))

# Ordenar as frases em ordem decrescente de importância e selecionar as principais
num_frases_resumo = 2
frases_importantes = sorted(frases_importantes, key=lambda x: x[1], reverse=True)[:num_frases_resumo]
resumo = '\n'.join([frase[0] for frase in frases_importantes])

print(resumo)
```

Neste exemplo, o código primeiro tokeniza o texto em frases e palavras, remove as *stopwords* e, em seguida, calcula a frequência de cada palavra no texto. Em seguida, ele identifica as frases mais importantes com base na pontuação total de suas palavras, e seleciona as duas frases mais importantes para criar um resumo extrativo do texto original. O resultado é um resumo com as frases mais importantes do texto original.

Também é possível mostrar a pontuação de cada frase no resumo:

```
for i, (frase, pontuacao) in enumerate(frases_importantes):
    print(f"({i+1}) {frase} (Pontuação: {pontuacao})")
```

Para mostrar as frases do resumo em destaque no texto original usando o módulo `HTML` do Python, você pode adicionar uma tag `HTML`, como `<mark>`, para cada frase do resumo e, em seguida, concatenar o texto original com essas tags. Aqui está um exemplo de como fazer isso:

```
# importando módulo HTML
from IPython.core.display import display, HTML

display(HTML(f'<h2>Sumarização Extrativa</h2>'))

# Adicionar tags HTML para destacar as frases do resumo no texto original
texto_com_destaque = texto_original
for frase, _ in frases_importantes:
  texto_com_destaque = texto_com_destaque.replace(frase, f"<mark>{frase}</mark>")

# Mostrar o resumo e as frases destacadas no texto original
display(HTML(f"<h3>Texto Original:</h3><p>{texto_com_destaque}</p>"))
display(HTML("<h3>Resumo:</h3>"))
for i, (frase, pontuacao) in enumerate(frases_importantes):
    display(HTML(f"<p>({i+1}) <mark>{frase}</mark> (Pontuação: {pontuacao})</p>"))
```

Aqui está um exemplo de código que utiliza a biblioteca `spaCy` para realizar a sumarização extrativa de um texto em português:

```
import spacy
from spacy.lang.pt.stop_words import STOP_WORDS
from collections import Counter

# Carregar o modelo em português do spaCy
nlp = spacy.load('pt_core_news_lg')

# Definir o texto a ser resumido
texto = """
A pandemia do novo coronavírus teve início em dezembro de 2019 na cidade de Wuhan, \
na China. Desde então, a doença se espalhou pelo mundo e foi declarada uma pandemia \
global pela Organização Mundial da Saúde (OMS) em março de 2020. A doença, conhecida \
como COVID-19, é causada pelo vírus SARS-CoV-2 e pode ser transmitida pelo ar ou pelo \
contato com superfícies contaminadas. Os sintomas mais comuns incluem febre, tosse e \
dificuldade para respirar. Para prevenir a disseminação da doença, as autoridades \
recomendam o uso de máscaras, a lavagem frequente das mãos e o distanciamento social.
"""

# Tokenizar o texto em frases e palavras usando o spaCy
doc = nlp(texto)
frases = [sent.text for sent in doc.sents]
palavras = [token.text for token in doc if not token.is_stop and token.is_alpha]

# Calcular a frequência de cada palavra no texto
frequencia_palavras = Counter(palavras)

# Calcular a pontuação de cada frase
pontuacao_frases = {}
for frase in frases:
    pontuacao = 0
    for palavra in nlp(frase):
        if palavra.text.lower() in frequencia_palavras:
            pontuacao += frequencia_palavras[palavra.text.lower()]
    pontuacao_frases[frase] = pontuacao

# Selecionar as frases mais importantes e gerar o resumo
num_frases_resumo = 2
frases_importantes = sorted(pontuacao_frases, key=pontuacao_frases.get, reverse=True)[:num_frases_resumo]
resumo = "\n".join(frases_importantes)

# Imprimir o resumo
print("Resumo:")
print(resumo)
```

Neste exemplo, o código utiliza a biblioteca `spaCy` para tokenizar o texto em frases e palavras em português, remover as *stopwords* e calcular a frequência de cada palavra. Em seguida, o código calcula a pontuação de cada frase com base na frequência das palavras que a compõem e seleciona as frases mais importantes para gerar o resumo.

```
import html
from IPython.core.display import HTML

# Destacando as frases do resumo no texto original em HTML
texto_html = ""
for frase in frases:
    if frase in frases_importantes:
        texto_html += "<mark>" + frase + "</mark> "
    else:
        texto_html += frase + " "

# Exibir o texto original com as frases do resumo destacadas em HTML usando a tag <mark>
display(HTML(html.unescape(texto_html)))
```

Existem diversas técnicas para fazer a sumarização, mas no próximo exemplo, iremos usar outra heurística simples.

Iremos assumir que, normalmente, uma frase que tem mais **entidades** e **substantivos** tem maior importância que outras sentenças. Usando essa lógica vamos ver um exemplo "simples" para cálculo de pontuação de importância. 

```
import spacy
from IPython.core.display import display, HTML

# Texto original
texto = """O Brasil é um país localizado na América do Sul, com uma área total de \
8.515.767,049 km² e uma população de mais de 213,8 milhões de pessoas, segundo dados \
de 2021. É o quinto maior país do mundo em área territorial e possui a maior parte da \
sua população concentrada na região sudeste do país. O Brasil é um país rico em recursos \
naturais, incluindo florestas tropicais, minerais e água doce. O Brasil também é um importante produtor \
de produtos agrícolas, como café, soja e carne bovina. Além disso, o país tem uma cultura \
rica e diversificada, com influências de Portugal, África, Europa e povos indígenas. \
O idioma oficial do Brasil é o português, e a religião predominante é o cristianismo, \
com uma grande população católica. O país é dividido em 26 estados e um Distrito Federal, \
cada um com sua própria cultura e tradições. O Brasil também é conhecido por seus esportes, \
como o futebol, que é o esporte mais popular do país. Apesar de seus muitos recursos e riquezas, o \
Brasil também enfrenta muitos desafios, como a desigualdade social, a pobreza e a corrupção. \
No entanto, o país continua a crescer e a se desenvolver, e tem um futuro promissor."""

# Carregando o modelo do spaCy em português
nlp = spacy.load("pt_core_news_lg")

# Processando o texto com o spaCy
doc = nlp(texto)

# Encontrando as entidades nomeadas e os substantivos mais importantes (que não são stopwords)
entidades_nomeadas = set([ent.text for ent in doc.ents if ent.label_ != ""])
substantivos = set([token.text for token in doc if token.pos_ == "NOUN" and token.is_stop == False])

# Destacando entidades nomeadas em vermelho e substantivos em azul
texto_formatado = texto
for entidade in entidades_nomeadas:
    texto_formatado = texto_formatado.replace(entidade, f"<span style='color:red'>{entidade}</span>")

for substantivo in substantivos:
    texto_formatado = texto_formatado.replace(substantivo, f"<span style='color:blue'>{substantivo}</span>")

texto_formatado = "<div style='max-width: 600px;'>" + texto_formatado + "</div>"
HTML(texto_formatado)
```

```
# Calculando a pontuação de cada frase
frases_pontuadas = []
for i, frase in enumerate(doc.sents):
    num_entidades = len(set([ent.text for ent in frase.ents if ent.label_ != ""]).intersection(entidades_nomeadas))
    num_substantivos = len(set([token.text for token in frase if token.pos_ == "NOUN" and not token.is_stop]).intersection(substantivos))
    pontuacao = num_entidades + num_substantivos
    frases_pontuadas.append((frase, i, pontuacao))

# Ordenando as frases pela pontuação em ordem decrescente
frases_pontuadas = sorted(frases_pontuadas, key=lambda x: x[2], reverse=True)

# Definindo o número de frases a serem incluídas no resumo
num_frases_resumo = 3

# Selecionando as frases mais importantes para o resumo
frases_importantes = [frase for frase, _, _ in frases_pontuadas[:num_frases_resumo]]
```

```
# Destacando as frases do resumo no texto original em HTML
texto_html = ""
for i, frase in enumerate(doc.sents):
    if frase in frases_importantes:
        texto_html += "<mark>" + frase.text + "</mark> "
    else:
        texto_html += frase.text + " "

# Exibir o texto original com as frases do resumo destacadas em HTML usando a tag <mark>
display(HTML(html.unescape(texto_html)))
```

```
from gensim.summarization import summarize
from IPython.core.display import HTML

# Definir a proporção do resumo em relação ao tamanho original do texto
SUMMARY_RATIO = 0.2

# Carregar o texto a ser resumido
text = '''
Artificial intelligence (AI) is rapidly changing the way we live and work. \
It is a branch of computer science that focuses on creating machines that can \
perform tasks that normally require human intelligence, such as visual perception, \
speech recognition, decision-making, and language translation. One of the most exciting \
applications of AI is in the field of healthcare. AI-powered systems are being used to \
diagnose diseases, develop new treatments, and even perform surgeries with greater accuracy \
and precision than human doctors. Another area where AI is having a big impact is in the realm \
of education. Intelligent tutoring systems are being developed to personalize learning for each \
student, while chatbots are being used to answer student queries in real-time. AI is also being \
used to enhance our transportation systems. Self-driving cars are already on our roads, \
and they are expected to greatly reduce the number of accidents caused by human error. In the \
field of finance, AI is helping banks and other financial institutions to better manage risk and \
identify fraud. AI-powered chatbots are also being used to provide customer service and answer queries. \
In the world of entertainment, AI is changing the way we consume and create content. \
Streaming services like Netflix use machine learning algorithms to recommend movies and \
TV shows to users, while AI-powered tools are being used to create music and artwork.
'''

# Fazer a sumarização do texto
summary = summarize(text, ratio=SUMMARY_RATIO)

# Formatar o resumo em uma tabela HTML
table = '<table><thead><tr><th>Resumo</th></tr></thead><tbody>'
table += f'<tr><td>{summary}</td></tr>'
table += '</tbody></table>'

# Exibir a tabela HTML com o resumo
display(HTML(table))
```

```
from IPython.core.display import display, HTML

# Separando as frases do texto original
sentences = text.split(".")

# Destacando as frases do resumo no texto original
highlighted_text = ""
for sentence in sentences:
    if sentence.strip() in summary:
        highlighted_text += "<mark>{}</mark>".format(sentence)
    else:
        highlighted_text += sentence

# Exibindo o texto original com o resumo destacado
display(HTML(highlighted_text))
```

**Sumarização Abstrativa**

Aqui está um exemplo de como usar o modelo `facebook/bart-large-cnn` para realizar a sumarização de um texto em inglês usando o `Hugging Face Transformers`.

`Hugging Face Transformers` é uma biblioteca de PLN de código aberto que fornece uma interface fácil de usar para usar e adaptar modelos pré-treinados para tarefas de PLN. Essa biblioteca é baseada em `PyTorch` e oferece suporte para uma ampla variedade de arquiteturas de modelo, incluindo transformers, bem como uma variedade de tarefas de PLN, como classificação de texto, sumarização, tradução e geração de linguagem natural. Com o `Hugging Face Transformers`, é fácil integrar modelos de última geração em seus aplicativos de PLN, mesmo se você não tiver experiência em *deep learning* (aprendizado profundo).

O modelo `facebook/bart-large-cnn` é um modelo de linguagem pré-treinado desenvolvido pelo Facebook. Ele utiliza a arquitetura do BART (*Bidirectional and Auto-Regressive Transformers*), que é um modelo que combina elementos do GPT (*Generative Pre-training Transformer*) e do BERT (*Bidirectional Encoder Representations from Transformers*).

O BART utiliza um processo de codificação e decodificação para gerar sequências de texto. Ele foi treinado com um grande conjunto de dados para predizer palavras ou frases faltantes em um texto. A versão "`large-cnn`" é uma variante com mais parâmetros do modelo original.

Esse modelo é particularmente eficiente para tarefas de geração de texto, como sumarização de texto e resposta automática a perguntas. Ele foi treinado em uma variedade de tarefas de linguagem natural, incluindo tradução de idiomas, preenchimento de lacunas em frases, e classificação de sentenças.

```
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

text = '''
Artificial intelligence (AI) is rapidly changing the way we live and work. \
It is a branch of computer science that focuses on creating machines that can \
perform tasks that normally require human intelligence, such as visual perception, \
speech recognition, decision-making, and language translation. One of the most exciting \
applications of AI is in the field of healthcare. AI-powered systems are being used to \
diagnose diseases, develop new treatments, and even perform surgeries with greater accuracy \
and precision than human doctors. Another area where AI is having a big impact is in the realm \
of education. Intelligent tutoring systems are being developed to personalize learning for each \
student, while chatbots are being used to answer student queries in real-time. AI is also being \
used to enhance our transportation systems. Self-driving cars are already on our roads, \
and they are expected to greatly reduce the number of accidents caused by human error. In the \
field of finance, AI is helping banks and other financial institutions to better manage risk and \
identify fraud. AI-powered chatbots are also being used to provide customer service and answer queries. \
In the world of entertainment, AI is changing the way we consume and create content. \
Streaming services like Netflix use machine learning algorithms to recommend movies and \
TV shows to users, while AI-powered tools are being used to create music and artwork.
'''

summary = summarizer(text)
print(summary)
```

Aqui está outro exemplo de como utilizar o modelo `facebook/bart-large-cnn` para gerar um resumo de um artigo sobre tecnologia:

```
from transformers import pipeline

article = """Technology is rapidly changing the world we live in. From smartphones to smart homes, we rely on technology more than ever before. One of the most exciting technological advancements in recent years is the development of artificial intelligence (AI). AI refers to the simulation of human intelligence in machines that are programmed to think and learn like humans.

The potential applications of AI are vast and varied. For example, self-driving cars use AI to make decisions based on real-time data from their surroundings. In the field of healthcare, AI is being used to diagnose diseases and develop personalized treatment plans for patients.

However, as with any new technology, there are also concerns about the potential risks and ethical implications of AI. For example, some people worry that AI could replace human jobs, or that it could be used to create autonomous weapons.

Despite these concerns, it seems clear that AI will continue to play an increasingly important role in our lives in the coming years. As such, it is essential that we approach this technology with caution and careful consideration of its potential benefits and drawbacks."""

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
summary = summarizer(article, max_length=100, min_length=30, do_sample=False)

print(summary[0]['summary_text'])
```

Esse código utiliza o modelo `facebook/bart-large-cnn` através do pipeline de sumarização do pacote transformers. O modelo é alimentado com um artigo sobre tecnologia e a função `summarizer` retorna um resumo do artigo com no máximo 100 tokens e no mínimo 30 tokens. O parâmetro `do_sample` foi configurado como `False`, o que significa que o modelo não irá gerar múltiplos resumos e selecionar o melhor, mas sim retornar o resumo mais provável.

**`Pegasus`**

`Google/PEGASUS-XSUM `é um modelo de linguagem baseado em `Transformer` que foi treinado em uma tarefa específica de sumarização de texto, conhecida como XSum (*Extreme Summarization*). A tarefa do XSum é criar uma breve descrição de uma notícia ou artigo, resumindo seu conteúdo em algumas frases.

O modelo `PEGASUS-XSUM` é uma versão do `PEGASUS` que foi treinado exclusivamente na tarefa XSum. O `PEGASUS` é uma abordagem para a criação de modelos de linguagem pré-treinados que foram treinados em várias tarefas de geração de texto, como sumarização, tradução e geração de texto.

O treinamento do `PEGASUS-XSUM` envolveu a utilização de um grande conjunto de dados de sumarização de texto para ensinar o modelo a identificar as informações mais importantes de um texto e a sintetizá-las em uma breve descrição. A partir daí, o modelo pode ser utilizado para gerar resumos de novos textos, que contenham as informações mais relevantes do texto original.

O `PEGASUS-XSUM` é considerado um dos melhores modelos de sumarização de texto atualmente disponíveis, pois consegue gerar resumos que contêm informações precisas e relevantes, ao mesmo tempo em que mantêm uma linguagem fluente e natural. Ele é amplamente utilizado em aplicações de PLN, como assistentes virtuais, sistemas de recomendação e motores de busca.

```
from transformers import pipeline

text = '''
Artificial intelligence (AI) is rapidly changing the way we live and work. \
It is a branch of computer science that focuses on creating machines that can \
perform tasks that normally require human intelligence, such as visual perception, \
speech recognition, decision-making, and language translation. One of the most exciting \
applications of AI is in the field of healthcare. AI-powered systems are being used to \
diagnose diseases, develop new treatments, and even perform surgeries with greater accuracy \
and precision than human doctors. Another area where AI is having a big impact is in the realm \
of education. Intelligent tutoring systems are being developed to personalize learning for each \
student, while chatbots are being used to answer student queries in real-time. AI is also being \
used to enhance our transportation systems. Self-driving cars are already on our roads, \
and they are expected to greatly reduce the number of accidents caused by human error. In the \
field of finance, AI is helping banks and other financial institutions to better manage risk and \
identify fraud. AI-powered chatbots are also being used to provide customer service and answer queries. \
In the world of entertainment, AI is changing the way we consume and create content. \
Streaming services like Netflix use machine learning algorithms to recommend movies and \
TV shows to users, while AI-powered tools are being used to create music and artwork.
'''

summarizer = pipeline("summarization", model = "google/pegasus-xsum")
summary = summarizer(text)

print(summary[0]['summary_text'])
```

**`T5-base`**

`T5-base` é um modelo de linguagem natural pré-treinado desenvolvido pelo Google. `T5` significa "*Text-to-Text Transfer Transformer*", e é baseado na arquitetura Transformer do modelo original do BERT.

O `T5` é um modelo altamente flexível, capaz de realizar diversas tarefas de PLN, como tradução, sumarização, resposta a perguntas e geração de texto. Ele é treinado em um conjunto de tarefas textuais amplas, utilizando um mecanismo de "treinamento por fim-a-fim", no qual o modelo recebe como entrada um texto e uma descrição da tarefa a ser realizada.

Uma das principais vantagens do `T5` é a sua capacidade de realizar tarefas de PLN sem a necessidade de ajustes finos específicos para cada tarefa, tornando-o uma escolha popular para aplicações em que há uma grande variedade de tarefas.

O `T5-base` é a versão menor do modelo `T5`, com um vocabulário menor e menos parâmetros. No entanto, ele ainda é capaz de realizar tarefas de PLN com bom desempenho, com a vantagem de exigir menos recursos computacionais.

```
text = """The tower is 324 metres (1,063 ft) tall, about the same height as an 81-storey building, \
and the tallest structure in Paris. Its base is square, measuring 125 metres (410 ft) on each side. \
During its construction, the Eiffel Tower surpassed the Washington Monument to become the tallest \
man-made structure in the world, a title it held for 41 years until the Chrysler Building in New York \
City was finished in 1930. It was the first structure to reach a height of 300 metres. Due to the \
addition of a broadcasting aerial at the top of the tower in 1957, it is now taller than the Chrysler \
Building by 5.2 metres (17 ft). Excluding transmitters, the Eiffel Tower is the second tallest free-standing \
structure in France after the Millau Viaduct."""

summarizer = pipeline("summarization", model = "t5-base", max_length=80)
summary = summarizer(text)

print(summary[0]['summary_text'])
```
