## **Modelagem de Tópicos**
---



A **Modelagem de Tópicos** é uma técnica de análise de dados que visa identificar tópicos ou temas latentes em um conjunto de documentos. É uma forma de análise de texto que utiliza algoritmos de aprendizado de máquina para extrair informações relevantes dos dados.

O processo começa com a coleta de um conjunto de documentos, que pode ser composto de artigos, posts de blog, e-mails, entre outros. Em seguida, são utilizados algoritmos de PLN para identificar palavras e frases relevantes em cada documento e mapeá-las para tópicos específicos. Esses tópicos podem ser definidos por meio de palavras-chave ou por uma descrição mais geral.

A modelagem de tópicos pode ser utilizada em diversas áreas, incluindo análise de mercado, jornalismo, pesquisa acadêmica, entre outras. Os resultados da modelagem de tópicos podem ser utilizados para identificar tendências, avaliar a reputação de marcas ou empresas, realizar análises de concorrência e até mesmo descobrir novas oportunidades de negócio.

A modelagem de tópicos é importante por vários motivos:

   * Identificação de tópicos relevantes: a modelagem de tópicos ajuda a identificar os tópicos mais relevantes em um conjunto de documentos. Isso permite que se tenha uma compreensão mais profunda do conteúdo e ajuda na análise e organização dos dados.

   * Descoberta de insights valiosos: ao identificar tópicos relevantes, a modelagem de tópicos pode revelar insights valiosos que podem ser utilizados para tomar decisões estratégicas e gerar oportunidades de negócio.

   * Eficiência no processamento de grandes volumes de dados: a modelagem de tópicos pode processar grandes volumes de dados rapidamente, o que torna o processo mais eficiente e econômico.

   * Melhoria na organização de documentos: a modelagem de tópicos pode ser utilizada para automatizar a classificação e organização de documentos, tornando o processo mais eficiente e reduzindo o risco de erros humanos.

   * Análise de tendências e comportamentos: a modelagem de tópicos pode ser utilizada para analisar tendências e comportamentos em um determinado período de tempo, o que pode ser útil em áreas como análise de mercado e previsão de comportamentos futuros.

   * Identificação de temas não esperados: a modelagem de tópicos pode revelar tópicos ou temas que não foram considerados anteriormente, o que pode levar a novas descobertas e insights.

   * Personalização de conteúdo: a modelagem de tópicos pode ser utilizada para personalizar o conteúdo para diferentes públicos, com base nos tópicos mais relevantes para cada grupo.

Embora a modelagem de tópicos seja uma técnica útil e eficaz para identificar tópicos relevantes em um conjunto de documentos, há alguns problemas que podem ocorrer:

   * Dependência da qualidade dos dados: a qualidade dos dados é um fator crucial na modelagem de tópicos. Se os dados estiverem incompletos, imprecisos ou contiverem informações irrelevantes, a precisão dos resultados pode ser afetada.

   * Dificuldade na interpretação dos resultados: a modelagem de tópicos pode gerar muitos tópicos e sub-tópicos, o que pode tornar a interpretação dos resultados um desafio. É importante ter uma compreensão clara dos algoritmos e métodos utilizados para garantir uma análise precisa e útil.

   * Sensibilidade aos parâmetros: a modelagem de tópicos é sensível a diferentes parâmetros, como número de tópicos e número de palavras-chave por tópico. A seleção desses parâmetros pode afetar significativamente os resultados da análise.

   * Dependência da língua e do contexto: a modelagem de tópicos pode ser influenciada pela língua e pelo contexto em que os documentos foram escritos. É importante levar em conta as nuances linguísticas e culturais para garantir uma análise precisa.

   * Limitações na representação de temas complexos: a modelagem de tópicos pode ter limitações na representação de temas complexos e multidisciplinares que não se enquadram facilmente em uma categoria específica.

Em resumo, a modelagem de tópicos pode enfrentar problemas relacionados à qualidade dos dados, interpretação dos resultados, sensibilidade aos parâmetros, dependência da língua e do contexto, e limitações na representação de temas complexos.

**Modelo LDA**

Existem diversas abordagens para a modelagem de tópicos, mas uma das mais populares é o modelo `LDA` (*Latent Dirichlet Allocation*). O modelo `LDA` é baseado em probabilidades e assume que cada documento é uma mistura de tópicos e que cada tópico é uma distribuição de palavras.

Aqui está um passo-a-passo básico de como é feita a modelagem de tópicos usando o modelo `LDA`:

   * Coletar os documentos: o primeiro passo é coletar um conjunto de documentos que serão usados para a modelagem de tópicos. Esses documentos podem ser artigos de jornal, postagens em redes sociais, comentários de clientes, entre outros.

   * Preprocessar os dados: antes de aplicar o modelo `LDA`, é necessário preprocessar os dados para remover stopwords (palavras comuns que não contribuem para o significado dos documentos), palavras com baixa frequência, palavras irrelevantes e aplicar outras técnicas de limpeza de texto, como tokenização, stemming e lematização.

   * Construir o vocabulário: depois de preprocessar os dados, é necessário construir um vocabulário a partir dos documentos. O vocabulário é uma lista de todas as palavras únicas que aparecem nos documentos e é usado para representar os documentos como vetores de palavras.

   * Representar os documentos como vetores: cada documento é representado como um vetor de palavras, onde cada dimensão representa uma palavra do vocabulário e o valor de cada dimensão representa a frequência ou a contagem de ocorrências dessa palavra no documento.

   * Aplicar o modelo `LDA`: com os documentos representados como vetores de palavras, o próximo passo é aplicar o modelo `LDA` para identificar os tópicos. O modelo `LDA` recebe como entrada os vetores de palavras dos documentos e retorna as distribuições de tópicos para cada documento e as distribuições de palavras para cada tópico.

   * Analisar os resultados: com o modelo `LDA` treinado, é possível analisar os resultados para entender melhor os tópicos identificados. É possível visualizar os tópicos e as palavras associadas a cada um e explorar os documentos que estão associados a cada tópico para entender melhor o conteúdo.

   * Ajustar os parâmetros: por fim, é possível ajustar os parâmetros do modelo `LDA` para melhorar a qualidade dos tópicos identificados. É possível ajustar o número de tópicos, o número de palavras associadas a cada tópico e outros parâmetros específicos do modelo.

É importante ressaltar que a modelagem de tópicos é um processo iterativo e que os resultados podem variar dependendo dos dados e dos parâmetros usados. Por isso, é importante experimentar diferentes abordagens e ajustar os parâmetros para obter os melhores resultados.

**`passes`**

No modelo LDA, a variável `passes` é um parâmetro que indica o número de vezes que o algoritmo LDA percorrerá o corpus de treinamento durante o processo de inferência dos tópicos. Cada passagem ou iteração (também chamada de `época`) consiste em duas fases: a fase de inferência e a fase de estimação dos parâmetros.

Durante a fase de inferência, o modelo LDA estima a distribuição de tópicos para cada documento do corpus, com base nas palavras que ocorrem no documento e na distribuição de tópicos em todo o corpus. Já na fase de estimação dos parâmetros, o modelo atualiza as probabilidades de tópicos e palavras, a fim de maximizar a verossimilhança do corpus.

Em geral, o parâmetro `passes` é ajustado com base no tamanho do corpus e na qualidade desejada dos tópicos inferidos. Quanto maior o número de `passes`, mais lenta será a inferência dos tópicos, mas a qualidade dos tópicos inferidos pode melhorar. Por outro lado, um número muito baixo de `passes` pode resultar em tópicos menos precisos.

Em resumo, o parâmetro `passes` controla o número de vezes que o modelo LDA irá iterar sobre o corpus de treinamento, ajustando gradualmente as distribuições de tópicos e palavras, até que os tópicos convergem para uma solução estável.

**`alpha`**

Na modelagem de tópicos com LDA, é comum que algumas palavras apareçam em mais de um tópico. Isso acontece porque as palavras podem ser relevantes para diferentes tópicos.

Por exemplo, considere que estamos modelando tópicos para notícias sobre esportes, e a palavra "gol" é uma das palavras mais relevantes em nosso corpus. É possível que a palavra "gol" esteja associada a tópicos como "futebol" e "hóquei no gelo", já que é uma palavra relevante para ambos os esportes.

Além disso, o LDA é um modelo probabilístico e, portanto, as palavras são associadas a tópicos com uma certa probabilidade. Ou seja, uma palavra pode ser mais fortemente associada a um determinado tópico, mas ainda ter uma probabilidade não nula de ser associada a outros tópicos.

No entanto, é possível controlar a frequência de palavras em diferentes tópicos através do parâmetro `alpha` no modelo LDA. Um valor baixo de `alpha` favorece tópicos mais concentrados em poucas palavras, enquanto um valor alto favorece tópicos mais distribuídos em várias palavras. Ajustar o valor de `alpha` pode ajudar a minimizar a presença de palavras em múltiplos tópicos.

Em `Gensim`, é possível passar o valor de `alpha` como parâmetro no método `LdaModel`. Se não for especificado, o valor padrão é `'1.0 / num_topics'`.

Uma palavra pode aparecer em dois tópicos com altas probabilidades porque as palavras muitas vezes são ambíguas e podem ter múltiplos significados que podem se relacionar com diferentes tópicos. Além disso, em alguns casos, a mesma palavra pode ser usada em diferentes contextos e, portanto, pode ser associada a diferentes tópicos com base em sua co-ocorrência com outras palavras no corpus de documentos.

No entanto, isso não é necessariamente um problema para a modelagem de tópicos, uma vez que o objetivo é identificar padrões e relacionamentos gerais entre as palavras e os tópicos, em vez de atribuir uma palavra exclusivamente a um único tópico. A sobreposição de palavras em diferentes tópicos é esperada e pode ser interpretada como uma indicação de que esses tópicos têm alguma relação ou semelhança conceitual.

**`Importante`**

O modelo LDA é um modelo probabilístico que envolve o uso de técnicas de amostragem para estimar as distribuições de tópicos e palavras em um corpus. Como resultado, a inferência do LDA é baseada em amostragem estocástica, ou seja, envolve a seleção aleatória de valores para parâmetros internos do modelo.

Devido a esse processo de amostragem estocástica, a inferência do LDA pode levar a diferentes resultados cada vez que o modelo é executado. Isso ocorre porque o modelo é sensível às condições iniciais e aos valores aleatórios selecionados durante a inferência. Pequenas variações nos valores iniciais ou na ordem das amostras podem levar a diferentes resultados finais.

Portanto, é comum observar variações nos resultados do modelo LDA, mesmo quando o mesmo corpus e os mesmos parâmetros são usados em diferentes execuções. Essa variação não é necessariamente um problema, mas pode ser considerada uma propriedade do modelo que reflete a incerteza na inferência dos tópicos.

Para minimizar a variação nos resultados, é possível definir a semente aleatória (*random seed*) para garantir que o modelo seja executado com os mesmos valores aleatórios a cada vez que for executado. Isso pode ser feito no `Gensim`, por exemplo, usando o parâmetro `random_state`.

Aqui está um exemplo simples de como fazer a modelagem de tópicos usando o modelo `LDA` em Python, utilizando a biblioteca `Gensim`:

```
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```
import gensim
from gensim import corpora
from gensim.models.ldamodel import LdaModel
from gensim.utils import simple_preprocess

# Define as stopwords em português
stop_words = set(stopwords.words('portuguese'))

# Define os documentos a serem analisados
documentos = [
    "A economia é o motor do desenvolvimento de um país.",
    "A economia mundial está em constante evolução.",
    "A crise econômica global afetou diversos setores da economia.",
    "O empreendedorismo é fundamental para a economia brasileira.",
    "O mercado financeiro é um importante indicador da economia.",
    "O setor de serviços é um dos principais motores da economia brasileira.",
    "A inovação tecnológica tem um grande impacto na economia global.",
    "A gestão financeira é essencial para o sucesso de um negócio.",
    "A sustentabilidade é um fator importante para o futuro da economia.",
    "A economia criativa é um setor em ascensão no mundo todo.",
    "ChatGPT é um modelo de linguagem natural desenvolvido pela OpenAI.",
    "ChatGPT é capaz de gerar textos em diversos estilos e temas.",
    "A qualidade das respostas geradas pelo ChatGPT é impressionante.",
    "ChatGPT é um avanço significativo na área de processamento de linguagem natural.",
    "A tecnologia do ChatGPT permite a interação com chatbots cada vez mais inteligentes.",
    "ChatGPT pode ser utilizado para gerar textos em diversos idiomas.",
    "ChatGPT é uma ferramenta poderosa para geração de conteúdo.",
    "ChatGPT utiliza a inteligência artificial para gerar textos em linguagem natural.",
    "ChatGPT é uma das mais avançadas ferramentas de geração de texto disponíveis atualmente.",
    "O ChatGPT é uma promessa para revolucionar a comunicação entre humanos e máquinas."
]

# Pré-processa os documentos
documentos_preprocessados = [simple_preprocess(doc, deacc=True) for doc in documentos]

# Remove as stopwords dos documentos
documentos_sem_stopwords = [[word for word in doc if word not in stop_words] for doc in documentos_preprocessados]

# Cria um dicionário com os documentos pré-processados
dicionario = corpora.Dictionary(documentos_sem_stopwords)

# Cria um corpus com base no dicionário e nos documentos pré-processados
corpus = [dicionario.doc2bow(doc) for doc in documentos_sem_stopwords]

# Define o número de tópicos a serem encontrados
num_topics = 2

# Executa a modelagem de tópicos usando o modelo LDA
modelo_lda = LdaModel(corpus, num_topics=num_topics, id2word=dicionario, passes=20)
#modelo_lda = LdaModel(corpus, num_topics=num_topics, id2word=dicionario, passes=20, random_state=17)
#modelo_lda = LdaModel(corpus, num_topics=num_topics, id2word=dicionario, passes=20, alpha=1)

# Imprime os tópicos e suas palavras-chave
for topico in modelo_lda.show_topics():
    print(topico)
```

**Conjunto de Dados [20 Newsgroups]**

`Newsgroups` é um conjunto de dados que consiste em cerca de 11.314 documentos de grupos de notícias de `20 tópicos` diferentes. Cada documento é rotulado com o nome do grupo de notícias ao qual pertence.

```
import pandas as pd

url = 'https://raw.githubusercontent.com/selva86/datasets/master/newsgroups.json'
df = pd.read_json(url)

print(df.target_names.unique())
```

```
import nltk
nltk.download('wordnet')
```

```
import gensim
from nltk.corpus import stopwords
from gensim.utils import simple_preprocess
from gensim.models import Phrases
from gensim.models.ldamodel import LdaModel
from gensim.corpora import Dictionary
from nltk.stem import WordNetLemmatizer

# define as stopwords em inglês
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

# cria um objeto lemmatizer
lemmatizer = WordNetLemmatizer()

# função para pré-processar o texto
def preprocess(text):
    # converte o texto em minúsculas e remove caracteres especiais
    text = text.lower()
    text = gensim.utils.simple_preprocess(text, deacc=True)

    # remove as stopwords e realiza a lematização
    text = [lemmatizer.lemmatize(w) for w in text if w not in stop_words]

    # cria bigramas e trigramas
    bigram = Phrases(text, min_count=5, threshold=100)
    trigram = Phrases(bigram[text], threshold=100)

    # retorna o texto pré-processado
    return trigram[bigram[text]]

# pré-processa o texto
df['clean_text'] = df['content'].apply(preprocess)

# cria o dicionário e o corpus
dictionary = Dictionary(df['clean_text'])
corpus = [dictionary.doc2bow(text) for text in df['clean_text']]

# define o número de tópicos
num_topics = 20

# executa o modelo LDA
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, passes=10)

# imprime os tópicos e as palavras mais relevantes de cada tópico
for i, topic in lda_model.show_topics(formatted=True, num_topics=num_topics, num_words=5):
    print('Topic {}: {}'.format(i, topic))
```

Uma forma de avaliar o modelo LDA é por meio de `Perplexity` (Perplexidade) e `Coherence Score` (Coerência).  A perplexidade mede a habilidade do modelo em prever novos dados, enquanto a coerência mede a interpretabilidade dos tópicos gerados.

Como regra geral para um bom modelo de LDA, a pontuação de perplexidade deve ser baixa, enquanto a coerência deve ser alta. Para perplexidade, o objeto `LdaModel` contém o método `log_perplexity` que pega um corpus como parâmetro e retorna a perplexidade correspondente. A biblioteca `Gensim` também possui uma classe `CoherenceModel` que pode ser usada para encontrar a coerência do modelo LDA.


```
print('\nPerplexity:', lda_model.log_perplexity(corpus))
```

```
from gensim.models import CoherenceModel

coherence_score_lda = CoherenceModel(model=lda_model, texts=df['clean_text'], dictionary=dictionary, coherence='c_v')
coherence_score = coherence_score_lda.get_coherence()

print('\nCoherence Score:', coherence_score)
```

A coerência varia de 0 a 1, sendo que valores mais próximos de 1 indicam um modelo mais coerente, ou seja, um modelo com tópicos mais distintos e semelhantes às palavras mais relevantes dentro de cada tópico. Por outro lado, valores mais baixos de coerência indicam um modelo com tópicos menos definidos e possivelmente mais semelhantes entre si. Vale ressaltar que os valores de coerência também podem variar dependendo do tipo de métrica de coerência escolhida.

Para calcular a coerência de cada tópico em um modelo LDA, podemos utilizar a biblioteca `gensim.models.coherencemodel` do `Gensim`. Vamos utilizar o modelo LDA que foi treinado anteriormente no conjunto de dados `Newsgroups` e calcular a coerência para cada um dos tópicos.

`u_mass` é uma das métricas de avaliação de coerência de tópicos em um modelo LDA. Essa métrica é baseada em uma abordagem não supervisionada que mede a relação semântica entre palavras usando um modelo estatístico. O valor varia de 0 a menos infinito, onde valores mais próximos de zero indicam uma maior coerência. Quanto mais negativo for o valor, menos coerente será o tópico.

```
# Calcular a coerência para cada tópico
coherence_model_lda = CoherenceModel(model=lda_model, corpus=corpus, coherence='u_mass')
coherence_scores = coherence_model_lda.get_coherence_per_topic()

# Exibir os scores de coerência para cada tópico em um gráfico de barras
import matplotlib.pyplot as plt

plt.bar(range(len(coherence_scores)), coherence_scores)
plt.xlabel('Tópicos')
plt.ylabel('Coerência')
plt.show()
```

O resultado é um gráfico de barras com os scores de coerência para cada um dos tópicos. Quanto maior o score, mais coerente é o tópico. Vale lembrar que a escala do score de coerência depende do método utilizado para calculá-lo, portanto é importante avaliar a coerência dos tópicos em relação aos demais tópicos do mesmo modelo.
