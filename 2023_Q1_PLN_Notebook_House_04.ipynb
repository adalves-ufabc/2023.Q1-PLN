```
from transformers import pipeline

# Define um modelo de Q&A
model = pipeline('question-answering', model='deepset/roberta-base-squad2')
```

```
def get_transcripts(df, episodes, characters=None):
    if characters:
        transcripts = df.loc[(df['character'].isin(characters)) & (df['episode'].isin(episodes)), 'transcript']
    else:
        transcripts = df.loc[df['episode'].isin(episodes), 'transcript']
    
    return ' '.join(transcripts.tolist())
```

```
def get_transcripts(df, episodes, characters=None):
    transcripts = ''
    for index, row in df.iterrows():
        if row['episode'] in episodes and (not characters or row['character'] in characters):
            transcripts += f"{row['character']}: {row['transcript']} "

    return transcripts.strip()
```

**`Episódio 1.01`**

```
transcripts = get_transcripts(df, [1.01], ['House', 'Cuddy', 'Wilson', 'Foreman', 'Chase', 'Cameron'])
```

```
# Define uma pergunta 
question = "Is the patient male or female?"

# Usa o modelo de Q&A para responder à pergunta
result = model(question=question, context=transcripts)

# Imprime a resposta
print(result)
```
