## **Similaridade de Textos**
---



**Similaridade de Textos** refere-se à medida de quão semelhantes são dois textos. Essa medida pode ser usada para comparar diferentes documentos, avaliar a originalidade de um texto ou detectar plágio.

Há muitas aplicações práticas para a similaridade de textos, incluindo:

   * Detecção de plágio: a similaridade de textos pode ser usada para detectar plágio em trabalhos acadêmicos, teses, artigos ou outros tipos de documentos. A comparação da similaridade entre um documento e uma base de dados de textos pode indicar se há evidência de plágio.

   * Recomendação de conteúdo: a similaridade de textos pode ser usada para recomendar conteúdo a usuários com base em suas preferências. Por exemplo, se um usuário gosta de um determinado tipo de artigo ou livro, a similaridade de texto pode ser usada para recomendar conteúdo semelhante.

   * Busca de informações: a similaridade de textos pode ser usada para ajudar na busca de informações. Por exemplo, se um usuário procura informações sobre um determinado tópico, a similaridade de texto pode ser usada para encontrar documentos semelhantes que possam fornecer informações relevantes.

   * Análise de sentimentos: a similaridade de textos pode ser usada para analisar sentimentos em comentários ou avaliações de usuários. A comparação da similaridade entre as opiniões de diferentes usuários pode indicar tendências ou opiniões gerais sobre um produto ou serviço.

   * Classificação de documentos: a similaridade de textos pode ser usada para classificar documentos em diferentes categorias com base em seu conteúdo. Por exemplo, se uma empresa tem um grande volume de documentos, a similaridade de texto pode ser usada para classificar esses documentos de acordo com suas categorias ou temas.

   * Identificação de fraudes: a similaridade de textos pode ser usada para identificar fraudes ou atividades suspeitas em documentos financeiros ou transações bancárias. A comparação da similaridade entre documentos pode ajudar a identificar possíveis irregularidades ou padrões suspeitos.

   * Análise de concorrência: a similaridade de textos pode ser usada para analisar a concorrência em um mercado ou setor específico. A comparação da similaridade entre os sites, publicidades e materiais promocionais de diferentes empresas pode ajudar a identificar tendências e estratégias de marketing.

   * Resumos automáticos: a similaridade de textos pode ser usada para gerar resumos automáticos de documentos longos. A comparação da similaridade entre diferentes seções de um documento pode ajudar a identificar as partes mais relevantes e importantes para incluir no resumo.

Algumas das principais vantagens da similaridade de textos são:

   * Eficiência: a análise de similaridade de textos pode ser realizada de forma rápida e eficiente, especialmente quando utilizamos técnicas computacionais. Isso significa que podemos processar grandes volumes de dados de forma mais eficiente do que se estivéssemos fazendo uma análise manual.

   * Precisão: a similaridade de textos pode fornecer resultados precisos e confiáveis. As técnicas utilizadas na análise de similaridade de textos são projetadas para fornecer uma medida objetiva de quão semelhantes são dois textos.

   * Flexibilidade: as técnicas de similaridade de textos podem ser aplicadas a diferentes tipos de documentos, desde artigos científicos até avaliações de produtos. Isso significa que podemos aplicar essas técnicas em diversas áreas de negócios e em diferentes contextos.

Apesar de suas vantagens, a similaridade de textos também apresenta alguns problemas e limitações, tais como:

   * Interpretação de contexto: as técnicas de similaridade de textos são baseadas em cálculos matemáticos, e muitas vezes não levam em consideração o contexto do texto. Isso significa que duas frases com palavras semelhantes podem ser consideradas semelhantes, mesmo que tenham significados completamente diferentes.

   * Complexidade do idioma: a similaridade de textos pode ser afetada pela complexidade do idioma utilizado nos documentos. Por exemplo, documentos técnicos e acadêmicos podem conter terminologia específica que pode afetar a precisão dos resultados de similaridade.

   * Variabilidade do texto: a similaridade de textos pode ser afetada pela variabilidade do texto. Por exemplo, uma pequena mudança em uma palavra ou frase pode levar a resultados diferentes de similaridade.

   * Dependência de dados: a similaridade de textos depende de um conjunto de dados de referência para comparar os documentos. Isso significa que a precisão da análise de similaridade pode ser afetada pela qualidade e tamanho desse conjunto de dados.

   * Dependência de parâmetros: as técnicas de similaridade de textos exigem que certos parâmetros sejam definidos para a análise. A seleção de parâmetros inadequados pode afetar a precisão dos resultados de similaridade.

   * Problemas de privacidade: a análise de similaridade de textos pode envolver o processamento de grandes volumes de dados sensíveis. Isso pode levantar questões de privacidade e segurança, especialmente se esses dados incluem informações pessoais ou confidenciais.

Existem várias técnicas para medir a similaridade de textos, algumas das quais são:

   * Similaridade de Hamming: essa técnica é usada para medir a distância entre duas strings de caracteres. Ela conta o número de posições em que duas strings diferem e divide pelo comprimento total da string.

   * Similaridade de Jaccard: essa técnica é baseada na análise da interseção e união de conjuntos de palavras em dois documentos. A similaridade de Jaccard é a razão da interseção de palavras pelo número total de palavras nos dois documentos.

   * Distância euclidiana: essa técnica calcula a distância euclidiana entre os vetores de palavras em dois documentos. Essa distância é calculada tomando a raiz quadrada da soma dos quadrados das diferenças entre as contagens de palavras em cada vetor.

   * Similaridade de cosseno: esta é uma das técnicas mais comuns usadas para medir a similaridade de textos. Ele usa o cosseno do ângulo entre dois vetores para calcular a semelhança entre eles. O vetor pode ser formado a partir da contagem de palavras em cada documento.   

Essas são apenas algumas das técnicas comuns usadas para medir a similaridade de textos. Dependendo do caso de uso, outras técnicas podem ser mais apropriadas.

**Similaridade de Hamming**

A **distância de Hamming** é uma métrica que mede a diferença entre duas sequências de igual comprimento, contando o número de posições em que essas sequências diferem. 

Para calcular a distância de Hamming entre duas sequências, é necessário comparar cada posição correspondente nas duas sequências e contar o número de posições em que os valores são diferentes. 

A distância de Hamming é uma métrica muito simples e útil para detectar erros em transmissões de dados, onde pode ser usada para detectar a presença de bits errados ou perdidos. 

```
def distancia_hamming(palavra1, palavra2):
    """Calcula a distância de Hamming entre duas palavras."""
    if len(palavra1) != len(palavra2):
        raise ValueError("As palavras precisam ter o mesmo comprimento.")
    return sum(c1 != c2 for c1, c2 in zip(palavra1, palavra2))

# Exemplo de uso:
palavra1 = "gato"
palavra2 = "gata"

distancia = distancia_hamming(palavra1, palavra2)
print(f"A distância de Hamming entre '{palavra1}' e '{palavra2}' é {distancia}.")
```

Neste exemplo, a função `distancia_hamming` recebe duas palavras `'palavra1'` e `'palavra2'` e calcula a distância de Hamming entre elas. A função compara cada letra nas duas palavras e conta o número de posições em que elas diferem. Em seguida, a distância de Hamming é retornada. O código também inclui um exemplo de uso da função, onde a distância de Hamming entre as palavras `"gato"` e `"gata"` é calculada e impressa.

A distância de Hamming é uma medida discreta que representa o número de posições em que as duas sequências são diferentes. No entanto, é possível calcular uma `"distância de Hamming normalizada"` dividindo a distância de Hamming pelo comprimento das sequências, o que resulta em um valor entre 0 e 1 que indica a proporção de posições em que as duas sequências são diferentes. Esse valor pode ser interpretado como uma medida de dissimilaridade ou `distância relativa` entre as duas sequências.

```
def hamming_distance_normalized(s1, s2):
    """Calcula a distância de Hamming normalizada entre duas sequências."""
    if len(s1) != len(s2):
        raise ValueError("As sequências precisam ter o mesmo comprimento.")
    d = sum(c1 != c2 for c1, c2 in zip(s1, s2))
    return d / len(s1)

# Exemplo de uso:
s1 = "abcd"
s2 = "abed"

distancia = hamming_distance_normalized(s1, s2)
print(f"A distância de Hamming normalizada entre '{s1}' e '{s2}' é {distancia:.2f}.")
```

**Similaridade de Jaccard**

A **Similaridade de Jaccard** é uma medida de similaridade entre dois conjuntos que avalia a proporção de elementos que são compartilhados entre eles. A similaridade de Jaccard é definida como a interseção do conjunto dividida pela união dos conjuntos.

A fórmula da similaridade de Jaccard é a seguinte:

> `J(A, B) = |A ∩ B| / |A ∪ B|`

onde A e B são dois conjuntos, A ∩ B é a interseção de A e B (ou seja, o conjunto de elementos que estão em ambos os conjuntos) e A ∪ B é a união de A e B (ou seja, o conjunto de elementos que estão em pelo menos um dos conjuntos).

O resultado da similaridade de Jaccard é um valor entre 0 e 1, em que 0 significa que não há elementos compartilhados entre os conjuntos e 1 significa que os conjuntos são idênticos. Quanto maior a similaridade de Jaccard, mais semelhantes são os conjuntos.

A similaridade de Jaccard é comumente usada em áreas como mineração de dados, aprendizado de máquina e PLN para comparar a similaridade entre conjuntos de itens. 

Aqui está um exemplo simples de como calcular a similaridade de Jaccard entre dois conjuntos:

```
def jaccard_similarity(set1, set2):
    """
    Calcula a similaridade de Jaccard entre dois conjuntos.
    """
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)

# Exemplo de uso
fruits1 = set(['maçã', 'banana', 'laranja', 'morango'])
fruits2 = set(['maçã', 'uva', 'kiwi'])

print("A similaridade de Jaccard entre os conjuntos é:", jaccard_similarity(fruits1, fruits2))
```

Aqui está um exemplo de como usar a similaridade de Jaccard para comparar duas frases:

```
def jaccard_similarity(text1, text2):
    """
    Calcula a similaridade de Jaccard entre duas frases.
    """
    # Transforma as frases em conjuntos de palavras únicas
    set1 = set(text1.split())
    set2 = set(text2.split())
    
    # Calcula a interseção e a união dos conjuntos de palavras
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    
    # Retorna a similaridade de Jaccard
    return len(intersection) / len(union)

# Exemplo de uso
text1 = "O gato está dormindo."
text2 = "O cachorro está acordado."

similarity = jaccard_similarity(text1, text2)
print("A similaridade de Jaccard entre as frases é:", similarity)
```

**Distância Euclidiana**

A **distância euclidiana** é uma medida de distância entre dois pontos em um espaço n-dimensional, que pode ser usada para calcular a distância entre dois vetores de características que representam dois textos. A distância euclidiana é calculada como a raiz quadrada da soma dos quadrados das diferenças entre as coordenadas dos vetores.

Para aplicar a distância euclidiana à similaridade de textos, primeiro é necessário representar cada texto como um vetor de características. Existem várias maneiras de fazer isso, como representar cada palavra como uma dimensão do vetor e a contagem de ocorrências de cada palavra como o valor da coordenada, ou usar técnicas de embedding para representar cada palavra como um vetor contínuo de dimensão fixa.

Suponha que temos dois textos representados como vetores de características `x` e `y`, cada um com `n` dimensões. A distância euclidiana entre os vetores `x` e `y` é dada pela seguinte fórmula:

> `d(x, y) = sqrt((x1 - y1)^2 + (x2 - y2)^2 + ... + (xn - yn)^2)`

Onde `x1, x2, ..., xn` são as coordenadas do vetor `x` e `y1, y2, ..., yn` são as coordenadas do vetor `y`. A raiz quadrada é usada para obter uma medida de distância positiva.



Para calcular a similaridade de textos usando a distância euclidiana, podemos usar a fórmula inversa da distância, que é dada por:

> `s(x, y) = 1 / (1 + d(x, y))`

onde `s(x, y)` é a similaridade entre os textos x e y. A fórmula acima retorna um valor entre 0 e 1, onde 0 indica que os textos são totalmente diferentes e 1 indica que os textos são idênticos. O uso da fórmula inversa da distância garante que quanto menor a distância euclidiana entre os vetores, maior será a similaridade entre os textos.

Segue abaixo um exemplo de código para calcular a similaridade de dois textos utilizando a distância euclidiana:

```
import math

def distancia_euclidiana(vetor1, vetor2):
    # Calcula a distância euclidiana entre os dois vetores
    distancia = math.sqrt(sum([(vetor1[i]-vetor2[i])**2 for i in range(len(vetor1))]))
    return distancia

def similaridade_euclidiana(texto1, texto2):
    # Calcula a similaridade entre os dois textos utilizando a distância euclidiana
    palavras = list(set(texto1.split() + texto2.split()))
    vetor1 = [texto1.split().count(palavra) for palavra in palavras]
    vetor2 = [texto2.split().count(palavra) for palavra in palavras]
    distancia = distancia_euclidiana(vetor1, vetor2)
    similaridade = 1 / (1 + distancia)
    return similaridade

# Exemplo de uso da função
texto1 = 'O gato comeu o rato'
texto2 = 'O cachorro comeu a ração'

similaridade = similaridade_euclidiana(texto1, texto2)
print('A similaridade entre os textos é: ', similaridade)
```

**Similaridade de Cosseno**

A **similaridade de cosseno** é uma medida de similaridade entre dois vetores, muito utilizada na área de PLN para comparar a similaridade entre documentos.

A ideia por trás da similaridade de cosseno é comparar o ângulo formado pelos vetores no espaço de n dimensões onde n é o número de termos distintos presentes nos documentos. A similaridade é calculada a partir do cosseno do ângulo entre esses vetores. Se o ângulo entre dois vetores for pequeno, então os vetores são considerados semelhantes e a similaridade de cosseno será próxima de 1. Se o ângulo for próximo a 90 graus, então os vetores são considerados dissimilares e a similaridade de cosseno será próxima de 0.



Em resumo, a similaridade de cosseno mede a semelhança entre dois vetores através do cosseno do ângulo formado entre eles. Quando aplicada a documentos, cada documento é representado como um vetor de termos, e a similaridade de cosseno é utilizada para comparar a similaridade entre esses vetores.

`TF-IDF` é uma medida de peso que atribui maior importância aos termos que aparecem com frequência em um documento, mas que aparecem raramente em outros documentos. `TF-IDF` é calculado multiplicando a frequência de um termo no documento (TF) pelo inverso da frequência desse termo em todo o conjunto de documentos (IDF).

A similaridade entre dois textos é então calculada utilizando a pontuação `TF-IDF` dos termos que aparecem em ambos os textos. Existem várias formas de calcular a similaridade usando `TF-IDF`, como a similaridade de cosseno, que mede a similaridade entre dois vetores de pesos de termos `TF-IDF`.

Aqui está um exemplo de código para calcular a similaridade de cosseno entre dois textos usando a biblioteca `scikit-learn` e a técnica `TF-IDF`:

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Textos de exemplo
text1 = "O Brasil é um país localizado na América do Sul"
text2 = "O Japão é um país localizado na Ásia"

# Vetorização dos textos usando TF-IDF
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform([text1, text2])

# Cálculo da similaridade de cosseno
cos_sim = cosine_similarity(tfidf[0], tfidf[1])

print(cos_sim)
```

Uma outra biblioteca que pode ser usada para cálculo de similaridade de cosseno é a biblioteca `numpy`. Aqui está um exemplo de código usando `numpy` para calcular a similaridade de cosseno entre dois vetores de texto:

```
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# exemplo de dois textos para calcular a similaridade de cosseno
text1 = "O Brasil é um país localizado na América do Sul"
text2 = "O Japão é um país localizado na Ásia"

# vetorização dos textos usando tf-idf
vectorizer = TfidfVectorizer()
vectorized_texts = vectorizer.fit_transform([text1, text2]).toarray()

# cálculo da similaridade de cosseno
cosine_similarity = np.dot(vectorized_texts[0], vectorized_texts[1]) / (np.linalg.norm(vectorized_texts[0]) * np.linalg.norm(vectorized_texts[1]))

print("Similaridade de cosseno entre os textos: ", cosine_similarity)
```

Para criar uma função que recomende títulos de filmes baseado na descrição semelhante usando similaridade de textos, podemos utilizar `TF-IDF` para representar cada descrição em um vetor numérico, e depois calcular a similaridade entre as descrições usando a distância do cosseno.

Para isso, podemos seguir os seguintes passos:

   1. Importar as bibliotecas necessárias: `pandas` para manipulação de dataframes, `scikit-learn` para o cálculo do `TF-IDF` e da similaridade de cosseno, e `numpy` para operações numéricas.

   2. Criar uma função que recebe como entrada um título de filme e um dataframe contendo títulos e descrições de filmes, e retorna uma lista com os títulos de filmes recomendados com base na similaridade de descrições.

   3. Na função, usar o título de entrada para selecionar a descrição correspondente no dataframe.

   4. Usar o `TF-IDFVectorizer` da biblioteca `scikit-learn` para calcular a representação `TF-IDF` de todas as descrições no dataframe.

   5. Calcular a matriz de similaridade de cosseno entre as descrições usando a função `cosine_similarity` da biblioteca `scikit-learn`.

   6. Selecionar as linhas da matriz de similaridade correspondentes à descrição do título de entrada e ordená-las em ordem decrescente de similaridade.

   7. Selecionar os títulos correspondentes às linhas selecionadas e adicioná-los a uma lista de recomendações.

   8. Retornar a lista de recomendações.

Segue abaixo uma possível implementação dessa função:

```
import nltk

nltk.download('stopwords')
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recomendar_filmes(titulo, df, n=5):
    # seleciona a descrição correspondente ao título de entrada
    descricao = df[df['titulo'] == titulo]['descricao'].values[0]

    stopwords = nltk.corpus.stopwords.words('portuguese')

    # calcula a representação TF-IDF de todas as descrições, desconsiderando as stopwords
    vectorizer = TfidfVectorizer(stop_words=stopwords)
    tfidf_matrix = vectorizer.fit_transform(df['descricao'])
    
    # calcula a matriz de similaridade de cosseno entre as descrições
    sim_matrix = cosine_similarity(tfidf_matrix)
    
    # seleciona as linhas correspondentes à descrição do título de entrada
    idx = df[df['titulo'] == titulo].index[0]
    sim_scores = list(enumerate(sim_matrix[idx]))
    
    # ordena as linhas em ordem decrescente de similaridade
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # seleciona os títulos correspondentes às linhas selecionadas e adiciona a uma lista de recomendações
    recommendations = []
    for i in range(1, n+1):
        title = df.iloc[sim_scores[i][0]]['titulo']
        sim = sim_scores[i][1]
        recommendations.append((title, sim))
    
    return recommendations
```

Essa função recebe três argumentos: o título de entrada, um dataframe contendo títulos e descrições de filmes, e o número de filmes a serem recomendados (padrão é 5).

```
# Exemplo de uso
recomendacoes = recomendar_filmes('O Poderoso Chefão', df)

# título do filme e similaridade
for recomendacao in recomendacoes:
  print(recomendacao)
```

**Scikit-learn**

`Scikit-learn` é uma biblioteca em Python de código aberto para aprendizado de máquina. Ela inclui uma ampla gama de algoritmos de classificação, regressão e clustering, bem como pré-processamento de dados e ferramentas de validação de modelo. A `Scikit-learn` é uma das bibliotecas mais populares para aprendizado de máquina em Python e é amplamente utilizada na indústria e em pesquisas acadêmicas.

Aqui está um exemplo básico de como usar o `TfidfVectorizer` e `cosine_similarity` do `scikit-learn` para calcular a similaridade entre textos:

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# lista de exemplos de textos
textos = [
    "Este é um exemplo de texto.",
    "Este é outro exemplo de texto.",
    "Este é um terceiro exemplo de texto."
]

# instanciando o vetorizador TF-IDF
tfidf = TfidfVectorizer()

# vetorizando os textos
tfidf_matrix = tfidf.fit_transform(textos)

# calculando a similaridade de cosseno entre todos os pares de textos
sim_matrix = cosine_similarity(tfidf_matrix)

# exibindo a matriz de similaridade
print(sim_matrix)
```

Aqui está um exemplo de como usar o `CountVectorizer` e o `cosine_similarity` do `scikit-learn` para encontrar a similaridade entre documentos. 

Inicialmente, criamos uma função para retornar uma lista de tuplas, onde cada tupla contém o documento mais similar e sua pontuação de similaridade.

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def encontrar_documentos_similares(input_doc, docs, n=3):
    # Vetorizando os documentos
    vectorizer = CountVectorizer()
    doc_vectors = vectorizer.fit_transform(docs)

    # Encontrando os documentos mais similares
    input_vector = vectorizer.transform([input_doc])
    similarity_scores = cosine_similarity(input_vector, doc_vectors)

    # Obtendo o índice dos documentos mais similares
    most_similar_docs_idx = similarity_scores.argsort()[0][::-1][:n]

    # Criando lista de tuplas com o documento mais similar e a pontuação de similaridade
    similar_docs = [(docs[idx], similarity_scores[0][idx]) for idx in most_similar_docs_idx]

    return similar_docs
```

Podemos usá-la da seguinte maneira:

```
docs = [
    "O Brasil é um país localizado na América do Sul",
    "A língua oficial do Brasil é o Português",
    "O carnaval do Brasil é conhecido em todo o mundo",
    "A Amazônia é a maior floresta tropical do mundo",
    "O futebol é o esporte mais popular do Brasil",
    "O Cristo Redentor é um dos pontos turísticos mais famosos do Brasil",
]

input_doc = "O Brasil é um país com uma grande diversidade cultural"
similar_docs = encontrar_documentos_similares(input_doc, docs, n=3)

print("Documentos mais similares:")
for doc, score in similar_docs:
    print("\n", doc)
    print("Pontuação de similaridade:", score)
```

`TfidfVectorizer` e `CountVectorizer` são duas classes do `scikit-learn` que permitem vetorizar textos, transformando-os em representações numéricas que podem ser usadas como entrada para algoritmos de aprendizado de máquina.

A principal diferença entre elas é que o `CountVectorizer` simplesmente conta o número de ocorrências de cada palavra em cada documento, enquanto o `TfidfVectorizer` pondera a importância de cada palavra em relação ao corpus de documentos.

O `TfidfVectorizer` calcula a frequência inversa do documento (IDF) para cada palavra, que é um indicador da raridade da palavra no corpus. Em seguida, o `TfidfVectorizer` multiplica a contagem de cada palavra em cada documento pelo IDF correspondente para obter uma pontuação `TF-IDF` para cada palavra.

Assim, palavras que aparecem frequentemente em um determinado documento, mas raramente em outros documentos do corpus, têm uma pontuação `TF-IDF` mais alta do que palavras que aparecem com a mesma frequência em todos os documentos. Isso pode ajudar a destacar palavras-chave importantes para cada documento, enquanto reduz o impacto de palavras comuns que não acrescentam muito valor semântico.

Ambas as técnicas, `TfidfVectorizer` e `CountVectorizer`, podem ser usadas para calcular a similaridade de textos. No entanto, a escolha entre as duas técnicas depende do objetivo do seu projeto e do tipo de texto que você está lidando.

O `CountVectorizer` é uma técnica mais simples que calcula a frequência de ocorrência de cada palavra em um texto. Ele é útil quando se deseja identificar a frequência de palavras em um texto, sem levar em conta a importância relativa dessas palavras em relação ao texto como um todo. O `CountVectorizer` pode ser usado para comparar documentos com base na frequência de ocorrência de palavras comuns.

O `TfidfVectorizer` é uma técnica mais avançada que leva em consideração a importância relativa das palavras em um texto. Ele calcula o peso de cada palavra com base na sua frequência de ocorrência no texto e na frequência em que a palavra ocorre em todo o conjunto de documentos. Isso significa que palavras que aparecem com frequência em um documento, mas não em outros, recebem uma pontuação maior do que palavras que aparecem com frequência em todos os documentos. O `TfidfVectorizer` é útil quando se deseja identificar palavras-chave importantes em um texto ou quando se deseja comparar documentos com base na importância relativa das palavras.

Em resumo, use o `CountVectorizer` se quiser uma comparação simples da frequência de palavras em textos. Use o `TfidfVectorizer` se quiser uma comparação mais sofisticada que leve em consideração a importância relativa das palavras nos textos.

```
import spacy

# Carregando o modelo do Spacy para a língua portuguesa
nlp = spacy.load('pt_core_news_lg')

# Definindo os textos para cálculo da similaridade
texto1 = "O Brasil é conhecido pelo seu futebol."
texto2 = "O futebol é um esporte muito popular no Brasil."

# Criando objetos Doc para cada texto
doc1 = nlp(texto1)
doc2 = nlp(texto2)

# Calculando a similaridade entre os textos
similaridade = doc1.similarity(doc2)

# Imprimindo a similaridade
print(f"A similaridade entre os textos é: {similaridade:.2f}")
```

Uma outra abordagem comum é utilizar o modelo pré-treinado de linguagem do `spaCy` para obter as representações vetoriais das palavras e, em seguida, calcular a similaridade de cosseno entre os vetores das frases.

Segue um exemplo de código que utiliza essa abordagem:

```
import spacy
from sklearn.metrics.pairwise import cosine_similarity
from typing import List, Tuple

def spacy_similarity(docs: List[str], input_doc: str, top_n: int = 3) -> List[Tuple[str, float]]:
    # Carregando o modelo de linguagem do Spacy para o idioma português
    nlp = spacy.load('pt_core_news_lg')

    # Processando os documentos com o modelo do Spacy e obtendo as representações vetoriais
    doc_vectors = [nlp(doc).vector for doc in docs]

    # Obtendo a representação vetorial do documento de entrada
    input_vector = nlp(input_doc).vector.reshape(1, -1)

    # Calculando a similaridade entre o documento de entrada e os documentos existentes
    similarity_scores = cosine_similarity(input_vector, doc_vectors)

    # Obtendo o índice dos documentos mais similares
    most_similar_docs_idx = similarity_scores.argsort()[0][::-1][:top_n]

    # Retornando a lista de documentos mais similares e seus scores de similaridade
    return [(docs[idx], similarity_scores[0][idx]) for idx in most_similar_docs_idx]
```

```
# Definindo alguns documentos
docs = [
    "O Brasil é um país localizado na América do Sul",
    "A língua oficial do Brasil é o Português",
    "O carnaval do Brasil é conhecido em todo o mundo",
    "A Amazônia é a maior floresta tropical do mundo",
    "O futebol é o esporte mais popular do Brasil",
    "O Cristo Redentor é um dos pontos turísticos mais famosos do Brasil",
]

# Testando a função
input_doc = "O Brasil é um país com uma grande diversidade cultural"
spacy_similarity(docs, input_doc)
```

O modelo `all-MiniLM-L6-v2` é uma versão pré-treinada da arquitetura MiniLM, desenvolvida pela Microsoft. Essa arquitetura é uma variação mais leve e compacta do modelo BERT, projetada para ser mais eficiente em termos de memória e tempo de treinamento. O modelo `all-MiniLM-L6-v2` foi treinado em um conjunto de dados multilíngue e pode ser usado para tarefas de PLN em vários idiomas.

```
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cosine

# Carregar o modelo all-MiniLM-L6-v2 pré-treinado
model = SentenceTransformer('all-MiniLM-L6-v2')

# Exemplo de textos a serem comparados
text1 = "O Brasil é um país localizado na América do Sul"
text2 = "A língua oficial do Brasil é o Português"

# Codificar os textos em vetores de alta dimensão usando o modelo
vector1 = model.encode(text1, convert_to_tensor=True)
vector2 = model.encode(text2, convert_to_tensor=True)

# Calcular a similaridade usando a distância do cosseno
similarity = 1 - cosine(vector1, vector2)

print("Similaridade:", similarity)
```

Aqui está um outro exemplo de código:

```
from sentence_transformers import SentenceTransformer, util

# Carregando o modelo pré-treinado
model = SentenceTransformer('all-MiniLM-L6-v2')

# 2 listas of sentenças
sentences1 = ['The cat sits outside',
             'A man is playing guitar',
             'The new movie is awesome']

sentences2 = ['The dog plays in the garden',
              'A woman watches TV',
              'The new movie is so great']

# Obter as representações vetoriais das frases
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(sentences2, convert_to_tensor=True)

# Calcular a similaridade 
cosine_scores = util.cos_sim(embeddings1, embeddings2)

# Imprimir o resultado
for i in range(len(sentences1)):
    print("{}      \t {} \t Score: {:.4f}".format(sentences1[i], sentences2[i], cosine_scores[i][i]))
```

O modelo `distilbert-base-nli-stsb-mean-tokens` é um modelo pré-treinado do `Transformer` que utiliza a arquitetura do BERT (*Bidirectional Encoder Representations from Transformers*) para criar representações de texto que são úteis para tarefas de PLN, como classificação de texto, análise de sentimento e similaridade de texto. Ele é pré-treinado usando o processo de Masked Language Model (MLM) e emparelhamento de texto, utilizando o conjunto de dados do GLUE (*General Language Understanding Evaluation*).

O modelo é específico para a tarefa de avaliação de similaridade de texto SemEval-2012 STS (S*emantic Textual Similarity*) e é treinado usando a métrica de correlação de Pearson. Ele é chamado de `distilbert-base-nli-stsb-mean-tokens` porque utiliza uma versão menor do modelo BERT chamada DistilBERT (uma versão mais leve do BERT) e as representações de texto são obtidas pela agregação da saída do token [CLS] (com a média) para as duas sequências de entrada (pares de frases) fornecidas.

```
from sentence_transformers import SentenceTransformer, util

# Carregando o modelo pré-treinado
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')

# Duas frases de exemplo
sentence1 = "O Brasil é um país localizado na América do Sul"
sentence2 = "O Japão é um país localizado na Ásia"

# Obter as representações vetoriais das frases
embeddings1 = model.encode(sentence1, convert_to_tensor=True)
embeddings2 = model.encode(sentence2, convert_to_tensor=True)

# Calcular a similaridade usando a função cosine_similarity
cosine_scores = util.pytorch_cos_sim(embeddings1, embeddings2)

# Imprimir o resultado
print("Similaridade: ", cosine_scores.item())
```
